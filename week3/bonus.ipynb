```python
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_csv('datasets/titles.csv')

dataset['release_year'].plot(kind = 'hist', bins = 85)
plt.xlabel("Year")
plt.ylabel("Film released")
plt.show()

highest = dataset[(dataset['release_year'] > 2017) & (dataset['release_year'] < 2022)]

highest.release_year.plot(kind = 'hist', bins = 7)
plt.xlabel("Year")
plt.ylabel("Film released")
plt.show()
```

```python
show_or_movie = dataset["type"].value_counts().iloc[0]
show_or_movie1 = dataset["type"].value_counts().iloc[1]

labels = 'SHOW ', 'MOVIE',
sizes = [show_or_movie, show_or_movie1]
explode = (0, 0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
